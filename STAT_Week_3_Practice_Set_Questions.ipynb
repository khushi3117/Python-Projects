```python
import numpy as np
from scipy import stats

# 1. Dataset
pot_mg = [
    151, 159, 168, 146, 129, 147, 149, 141, 150, 158, 119, 125, 139, 154, 163,
    156, 132, 137, 126, 152, 127, 135, 138, 145, 128, 148, 153, 124, 170, 134,
    157, 164, 144, 160, 161
]

# 2. Parameters
mu_0 = 148       # Hypothesized mean
alpha = 0.05     # Significance level

# 3. Descriptive Statistics
sample_mean = np.mean(pot_mg)
sample_std = np.std(pot_mg, ddof=1)  # Use ddof=1 for sample standard deviation
n = len(pot_mg)

# 4. Perform One-Sample T-Test (Left-Tailed)
# We use alternative='less' because the claim is that the mean is LESS than 148
t_stat, p_value = stats.ttest_1samp(pot_mg, mu_0, alternative='less')

# 5. Output Results
print(f"Sample Mean: {sample_mean:.2f} mg")
print(f"Sample Standard Deviation: {sample_std:.2f} mg")
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# 6. Conclusion
if p_value < alpha:
    print("Decision: Reject the Null Hypothesis (Support the claim).")
else:
    print("Decision: Fail to Reject the Null Hypothesis (Do not support the claim).")
```

```python
# It is two taild test, given values.
from scipy import stats
import math
n=60
# null hypothesis=the average length of rods is 26cm, mean mu=26cm
# altenative hypothesis =the avrege length of the rod is diffrent from 26cm mu not is 26
mu0=26     # Hypothesized population mean
p_std=2.4   #Population standard deviation
x=24.8   #sample mean
confidence_level=0.95
alpha=1-0.95
# calculate the z-score value
z=(x-mu0)/(p_std / math.sqrt(n))
print("z is:",z)
# calculate the p_value two tailed test
# Since it's two-tailed, we find the area in one tail and multiply by 2
p_value=2*stats.norm.cdf(z)
print("p value is:",p_value)
# determine the critical z values
#For a 95% confidence level, alpha/2 = 0.025
z_critical = stats.norm.ppf(alpha/2)
print("z critical values:",z_critical)
# calculate the margin error
ME=z_critical*p_std
print("margin error:",ME)

confidence_interval=(x-ME,x+ME)

print("confidence interval for upper bound and lower bpund:",confidence_interval)
# Decision Logic

if abs(z)>z_critical:
  print("\nDecision: Reject the Null Hypothesis (H0)")
  print("Conclusion: The average length is significantly different from 26 cm.")
else:
  print("\nDecision: Fail to Reject the Null Hypothesis (H0)")
  print("Conclusion: There is no significant difference from 26 cm.")
```
