## STACKING-->COMBINE ONE OR MORE TABLE TO FORM A SINGLE ARRAY WITH HIGHER DIMENSION

```python
num_array_1 = np.array([[1, 2, 3], [4, 5, 6]])
num_array_2 = np.array([[7, 8, 9], [10, 11, 12]])
print(num_array_1)
print(num_array_2)
```

```python
np.stack([num_array_1,num_array_2])
```

```python
np.stack([num_array_1,num_array_2],axis=0)
```

```python
np.stack([num_array_1,num_array_2],axis=1)
```

```python
np.hstack([num_array_1,num_array_2])
```

```python
np.vstack([num_array_1,num_array_2])
```

```python
np.dstack([num_array_1,num_array_2])
```
